```python
distances = [[0.0 for _ in range(len(list_of_food_trucks))] for _ in range(len(list_of_food_trucks))]

for i in range(len(list_of_food_trucks)):
    if list_of_food_trucks[i] == 'IUPUI':
        latitude_1 = location[0]
        longitude_1 = location[1]
    else:
        latitude_1 = details[list_of_food_trucks[i]]['Location']['lat']
        longitude_1 = details[list_of_food_trucks[i]]['Location']['lng']
    for j in range(len(list_of_food_trucks)):
        if list_of_food_trucks[j] == 'IUPUI':
            latitude_2 = location[0]
            longitude_2 = location[1]
        else:
            latitude_2 = details[list_of_food_trucks[j]]['Location']['lat']
            longitude_2 = details[list_of_food_trucks[j]]['Location']['lng']
        distance = math.sqrt((latitude_1 - latitude_2)**2 + (longitude_1 - longitude_2)**2)
        distances[i][j] = distance

distances
```

```python
list_of_unique_cuisines = set([value for key, value in cuisines.items()])
list_of_unique_cuisines.remove('Various')
print(list_of_unique_cuisines)
list_of_unique_truck_names = set(list_of_food_trucks[:])
print(list_of_unique_truck_names)
```

```python
import heapq
priority_queue = []

def multi_priority(item):
    return (item[0], -item[1])

final_trucks_list = []
for i in range(2):
    temp_list = ['IUPUI']
    from_places = None
    for j in range(3):
        if j == 0:
            from_places = 'IUPUI'
        list_of_unique_truck_names.remove(from_places)
        index = list_of_food_trucks.index(from_places)
#         print(index)
        for k in range(21):
            if k != index and k!=0:
                heapq.heappush(priority_queue, (distances[index][k], details[list_of_food_trucks[k]]['Rating'], k))
        while priority_queue:
            best_choice = heapq.heappop(priority_queue)[2]
            if (list_of_food_trucks[best_choice] in list_of_unique_truck_names) and (details[list_of_food_trucks[best_choice]]['Cuisine'] in list_of_unique_cuisines):
                temp_list.append(list_of_food_trucks[best_choice])
                from_places = list_of_food_trucks[best_choice]
                print(from_places)
                list_of_unique_cuisines.remove(details[list_of_food_trucks[best_choice]]['Cuisine'])
                break
                #list_of_unique_truck_names.remove(list_of_food_trucks[best_choice])
#     print(i)
    final_trucks_list.extend(temp_list)
    temp_list.clear()
    list_of_unique_truck_names.add('IUPUI')
print(final_trucks_list)
```
